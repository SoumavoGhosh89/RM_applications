# Comparison plots:

```python
#prepare pretty plotting:
from matplotlib import rcParams
from matplotlib.colors import LogNorm
rcParams['font.size'] = 16
rcParams['axes.linewidth'] = 1.5
rcParams['xtick.major.width'] = 1.5
rcParams['xtick.minor.width'] = 1.5
rcParams['ytick.major.width'] = 1.5
rcParams['ytick.minor.width'] = 1.5
```

```python
fig = plt.figure(figsize=(6,15))

axislist = [fig.add_subplot(6,1,ii+1) for ii in range(6)]
datarange = numpy.array([[230.,250.],[2.0,3.5],[0.,0.6],[-1.5,-0.5],[-1.95,-1.75],[-2.4,-1.5]])
for ii in range(6):
    ax = axislist[ii]
    ax.set_ylim(datarange[ii])
    ax.set_xlim([-0.5,1.5])
    ax.set_xticks([])
    
ax = axislist[-1]
ax.set_xticks([0.,1.])
ax.set_xticklabels(['A','B'])

for ii,datasetname,testname,color in zip(range(2),
                                   ['mock_data_exampleA_py3','mock_data_exampleB_py3'],
                                   ['test1py3','test0'],['purple','hotpink']):
    pos_in_plot = ii
    oplot_violines(datasetname,pos_in_plot,
                                testname=testname,
                                method='MCMC',
                                quantities_to_plot=None,
                                list_of_axis_objects=axislist,
                                width=0.5,
                                analysis_output_filename=None,
                                color=color,
                                columns=1,
                                this_column=1,
                                sigma_levels=False,
                                plot_violins=True,
                                plot_true_values=True,
                                show_labels=True,
                                burnin_steps=100,
                                datapath="../data/",
                                outpath="../out/",
                                Gaussian_fit=False
                  )
                   
plt.tight_layout()
plt.show()
```
